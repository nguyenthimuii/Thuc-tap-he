```python
# Drop 4 cột có giá trị không đổi là EmployeeCount,Over18,StandardHours,EmployeeNumber
df_clean = df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
df_clean.head(25)
```

```python
df_clean.shape
```

```python
# Tìm các cột có giá trị bị thiếu
columns_with_missing_values = df_clean.columns[df_clean.isnull().any()]

# Đếm số giá trị thiếu trong từng cột
missing_value_counts = df_clean[columns_with_missing_values].isnull().sum()
print("Các cột có giá trị bị thiếu:")
print(columns_with_missing_values)
print("\nSố giá trị thiếu trong từng cột:")
print(missing_value_counts)
```

```python
# Xác định kiểu dữ liệu của cột
column_data_types = df_clean.dtypes
# Đếm các cột số và phân loại
numerical_count = 0
categorical_count = 0
for column_name, data_type in column_data_types.items():
    if np.issubdtype(data_type, np.number):
        numerical_count += 1
    else:
        categorical_count += 1
print(f"Trong bộ dữ liệu có :")
print(f"Có {numerical_count} cột biến số")
print(f"Có {categorical_count} cột biến phân loại")
```
